```python
import skimage
import skimage.filters
import matplotlib.pyplot as plt
import numpy as np
```

```python

def show_image(image, title='Image'):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
#i=skimage.data.cell()
from skimage import data
i=data.cell()
show_image(i)
i=skimage.img_as_float(i)
a=.5
i2=i*a
plt.imshow(i2,'gray',vmin=0,vmax=1)
plt.show()
```

```python
from scipy.ndimage.filters import uniform_filter
i_moy=uniform_filter(i,size=(150,150))
plt.imshow(i_moy,'gray')
plt.show()
```

```python
from scipy.ndimage.filters import median_filter
i_med=median_filter(i,size=(150,150))
plt.imshow(i_med,'gray')
plt.show()
```

```python
i_gauss=skimage.filters.gaussian(i,15)
plt.imshow(i_gauss,'gray')
plt.show()
```

```python
i_sobel=skimage.filters.sobel(i)
plt.imshow(i_sobel,'gray')
plt.show()
```

```python
i_scharr=skimage.filters.scharr(i)
plt.imshow(i_scharr,'gray')
plt.show()
```

```python
i_hess=skimage.filters.hessian(i,mode="constant")
plt.imshow(i_hess,'gray')
plt.show()
```
