```python
import skimage
import skimage.filters
import matplotlib.pyplot as plt
import numpy as np
```

```python
from skimage import data
i=skimage.data.camera()
plt.imshow(i,'gray')
plt.show()
```

```python
from scipy.signal import convolve2d
```

```python
taille=15
h=np.ones((taille,taille))/taille**2
i_moy=convolve2d(i,h,mode='same')
plt.imshow(i_moy,'gray')
plt.show()
```

```python
k=15
sigma=7
x=np.arange(-k,k+1)
[i,j]=np.meshgrid(x,x)
h=1/(2*np.pi*sigma**2)*np.exp(-(i**2+j**2)/(2*sigma**2))
h=h/np.sum(h)
```

```python
plt.imshow(h)
plt.show()
print(h.shape)
```

```python
i=skimage.data.camera()
i_gauss2=convolve2d(i,h,mode='same')
plt.imshow(i_gauss2,'gray')
plt.show()
```

```python
i=skimage.data.camera()
h=1/16 * np.array([[1,2,1],[2,4,2],[1,2,1]])
i_gauss=convolve2d(i,h,mode='same')
plt.imshow(i_gauss,'gray')
plt.show()
```

```python
h=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
ix=convolve2d(i,h,mode='same')
plt.imshow(ix,  'gray')
plt.show()
```

```python
h=h.transpose()
iy=convolve2d(i,h,mode='same')
plt.imshow(iy,'gray')
plt.show()
```

```python
gradient= np.sqrt(ix**2 +  iy**2)
plt.imshow(gradient,'gray')
plt.show()
```

```python
hlab = np.array([[0,1,0],[1,-4,1],[0,1,0]])
laplacien = convolve2d(i,hlab,mode='same')
plt.imshow(laplacien,'gray')
plt.show()
```

```python
import skimage as ski
import skimage.io as io
import skimage.util as util
import matplotlib.pyplot as plt
import numpy as np
from skimage import color
```

```python
I=io.imread("parapente.jpg")
a = color.rgb2gray(I)
x=ski.img_as_float(a)

```

```python
plt.figure(figsize=(8,8))
plt.imshow(a,'gray')
plt.show()
```

```python
J= util.random_noise(a,mode="s&p",amount=0.1)
x1=10
y1=10
x2=250
y2=180
K=J[y1:y2,x1:x2]
plt.figure(figsize=(8,8))
plt.imshow(J,'gray')
plt.plot([x1,x2,x2,x1,x1],[y1,y1,y2,y2,y1],'g')
plt.show()
```

```python
r,bins,patches=plt.hist(K.ravel())
plt.show()
```

```python
M, N = K.shape[0], K.shape[1]
N0=np.sum(K==0)
N1=np.sum(K==1)
prop=(N0+N1)/(M*N)
print("Nombre de pixels : %d  -  Nombre de pixels  0 : %d  -  Nombre de pixels  1 : %d "%(M*N,N0,N1))
print("proportion de pixels dfaillante : %.3f" %(prop))
```
