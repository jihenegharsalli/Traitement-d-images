```python
import matplotlib.pyplot as plt
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
from skimage import color
grayscale = color.rgb2gray(rocket_image)
show_image(grayscale, "Grayscale")
```

```python
flipped_seville = plt.imread('Sevilleup.jpg')
show_image(flipped_seville)
```

```python
# Flip the image in up direction
vertically_flipped = np.flipud(flipped_seville)
show_image(vertically_flipped, 'Vertically flipped image')
```

```python
# Flip the image in left direction
horizontally_flipped = np.fliplr(vertically_flipped)
show_image(horizontally_flipped, 'Horizontally flipped image')
```

```python
# Import the otsu threshold function
from skimage.filters import threshold_otsu
```

```python
bw = plt.imread('bw.jpg')
grayscale = color.rgb2gray(bw)
show_image(grayscale)
thresh = threshold_otsu(grayscale)
binary_global = grayscale > thresh
show_image(bw, 'Original')
show_image(binary_global, 'Global thresholding')
```
